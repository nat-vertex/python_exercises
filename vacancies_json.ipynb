```python
import requests    
import json          
import time         
import os            
import pandas as pd  


class Params:
    def __init__(self, arguments={}):
        self.__arguments = arguments
    
    @property   
    def arguments(self):
        return self.__arguments
    @arguments.setter
    def arguments(self, arguments):
        self.__arguments = arguments

class PullData:
    def __init__(self, file_path, file_name, obj_params, address):
        self.file_path = file_path
        self.file_name = file_name 
        self.address = address
        self.obj_params = obj_params
        self.flag = True
        
    def create_path(self):
        os.makedirs(self.file_path, exist_ok=True)
        self.file_name = self.file_path + self.file_name + '.json'
        self.flag = False 
        
    def pull_data(self):
        if self.flag:
            self.create_path()
        json_obj = requests.get(self.address, params=self.obj_params.arguments).json()
        #print(json_obj)
        with open(self.file_name, "a", encoding="utf-8") as f:
            json.dump(json_obj, f, ensure_ascii=False, indent=4)
            
    def get_pages(self): 
        return requests.get(self.address, params=self.obj_params.arguments).json()['pages']
    
    

params_req = Params()
params_req.arguments = {
        'text': 0, 
        'page': 0,     
        'search_field': 'name',
        'per_page': 100      
        }   
text_names = ['"data engineer"', 
              '"Инженер данных"', 
              '"data инженер"', 
              '"дата инженер"']

for text_name in text_names:
    file_path = './hh_api_files/'
    file_name = 'all_entries3'
    address = 'https://api.hh.ru/vacancies'
    
    file_with_vacancies = PullData(file_path, file_name, params_req, address)
    params_req.arguments['page'] = 0
    params_req.arguments['text'] = text_name
    n = file_with_vacancies.get_pages()
            
    for page in range(0, n): 
        params_req.arguments['page'] = page
        file_with_vacancies.pull_data()
              
```
